# Fungsi Nilai Asli Integral

```python
import numpy as np
import sympy as sp

def hitung_integral(f, a, b, n=1000):
    """
    Menghitung integral numerik menggunakan metode trapesium.
    
    Parameters:
        f (function): Fungsi yang akan diintegralkan
        a (float): Batas bawah integral
        b (float): Batas atas integral
        n (int): Jumlah titik partisi (default: 1000)
        
    Returns:
        float: Hasil aproksimasi integral
    """
    x = np.linspace(a, b, n)
    y = f(x)
    return np.trapezoid(y, x)
```

# Metode Trapesium
Metode trapesium merupakan salah satu metode numerik dalam menyeles aikan integral yang membagi interval integrasi menjadi beberapa trapesium dengan lebar yang sama, kemudian menghitung luas setiap trapesium dan menjumlahkannya untuk mendapatkan pendekatan nilai integral.<br>
Aturan trapesium bekerja dengan cara mengaproksimasikan daerah di bawah grafik fungsi 
gaya tampilan f(x)sebagai trapesium dan menghitung luasnya. Maka dapat disimpulkan bahwa: <br>
$\displaystyle{\int_{{a}}^{{b}}} f{{\left({x}\right)}}{\left.{d}{x}\right.}={h}\cdot{\left[ f{{\left({x}_{{0}}\right)}}+{2} f{{\left({x}_{{1}}\right)}}+{2} f{{\left({x}_{{2}}\right)}}+{2} f{{\left({x}_{{3}}\right)}}+\ldots+{2} f{{\left({x}_{{{n}-{1}}}\right)}}+ f{{\left({x}_{{n}}\right)}}\right]}$<br>
Dan dapat disederhanakan menjadi:<br>
$\displaystyle{\int_{{a}}^{{b}}} f{{\left({x}\right)}}{\left.{d}{x}\right.}={h}\cdot{\left[ f{{\left({x}_{{0}}\right)}}+{2}\cdot{\sum_{{{i}={1}}}^{{{n}-{1}}}} f{{\left({x}_{{i}}\right)}}+ f{{\left({x}_{{n}}\right)}}\right]}$

```python
import numpy as np
import matplotlib.pyplot as plt

def Trapesium(f, a, b, h):
    x = [a]
    i = 0
    
    while x[-1] < b:
        a += h
        if a > b:
            break
        x.append(a)
    x = np.array(x)
    y = f(x)
    n = len(x)
    
    hasil = y[0] + y[-1]
    for i in range(1, n-1):
        hasil += 2 * y[i]
    
    return x, y, hasil

# Definisikan fungsi dan parameter
f = lambda x: np.sin(2*x)
a = 0
b = 5
h = 1

x_trap, y_trap, integral = Trapesium(f, a, b, h)
galat = abs((exact - integral)/exact)*100

# Generate kurva halus untuk fungsi asli
x_smooth = np.linspace(a, b, 1000)
y_smooth = f(x_smooth)

# Plot grafik
plt.figure(figsize=(10, 6))

# Plot fungsi asli
plt.plot(x_smooth, y_smooth, 'b-', label='Fungsi Asli: $f(x) = 2x$')

# Plot garis trapesium (menghubungkan titik-titik sampel)
plt.plot(x_trap, y_trap, 'ro--', label='Pendekatan Trapesium')

# Arsir daerah di bawah trapesium
plt.fill_between(x_trap, y_trap, 0, color='gray', alpha=0.3, label='Daerah Integral')

# Tambahkan label dan judul
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(f'Integral dengan Metode Trapesium \nLebar Langkah (h) = {h} \nHasil numberik: {integral:.2f}, nilai sejati: {exact:.2f} \nGalat: {galat:.2f}%')
plt.legend()
plt.grid(True)
plt.show()
```

# Metode Simpson 1/3
Integral metode Simpson 1/3 adalah metode numerik yang menggunakan polinomial berderajat dua untuk mendekati fungsi dan mencari nilai perkiraan integral. Metode ini berguna ketika tidak mungkin menerapkan teknik integrasi lain. <br>
Disimpulkan bahwa: <br>
$\displaystyle{\int_{{a}}^{{b}}} f{{\left({x}\right)}}{\left.{d}{x}\right.}={h}\cdot{\left[ f{{\left({x}_{{0}}\right)}}+{4} f{{\left({x}_{{1}}\right)}}+{2} f{{\left({x}_{{2}}\right)}}+{4} f{{\left({x}_{{3}}\right)}}+{2} f{{\left({x}_{{4}}\right)}}+\ldots+ f{{\left({x}_{{n}}\right)}}\right]}$<br>
Dan dapat disederhanakan menjadi:<br>
$\displaystyle{\int_{{a}}^{{b}}} f{{\left({x}\right)}}{\left.{d}{x}\right.}={h}\cdot{\left[ f{{\left({x}_{{0}}\right)}}+{4}\cdot{\sum_{{{i}={G}{a}{n}{j}{i}{l}}}^{{{n}-{1}}}} f{{\left({x}_{{i}}\right)}}+{2}\cdot{\sum_{{{i}={G}{e}{n}{a}{p}}}^{{{n}-{2}}}} f{{\left({x}_{{i}}\right)}}+ f{{\left({x}_{{n}}\right)}}\right]}$

```python
import numpy as np
import matplotlib.pyplot as plt

def Simpson_13(f, a, b, h):
    x = [a]
    i = 0
    
    while x[-1] < b:
        a += h
        if a > b:
            break
        x.append(a)
        
    x = np.array(x)
    y = f(x)
    n = len(x)
    
    hasil = y[0] + y[-1]
    for i in range(1, n-1):
        if i % 2 != 0:
            hasil += 4 * y[i]
        elif i % 2 == 0:
            hasil += 2 * y[i]
    hasil = (h / 3) * hasil
    
    return x, y, hasil

# Definisikan fungsi dan parameter
f = lambda x: np.sin(2*x)
a = 0
b = 6
h = 1

x_simp, y_simp, integral = Simpson_13(f, a, b, h)
galat = abs((exact - integral)/exact)*100

# Generate kurva halus untuk fungsi asli
x_smooth = np.linspace(a, b, 1000)
y_smooth = f(x_smooth)

# Persiapkan titik-titik parabola untuk visualisasi
x_composite = np.array([])
y_composite = np.array([])

# Fit parabola untuk setiap 3 titik
for i in range(0, len(x_simp)-1, 2):
    x0, x1, x2 = x_simp[i], x_simp[i+1], x_simp[i+2]
    y0, y1, y2 = y_simp[i], y_simp[i+1], y_simp[i+2]
    
    # Koefisien parabola
    coeffs = np.polyfit([x0, x1, x2], [y0, y1, y2], 2)
    
    # Generate titik parabola
    x_para = np.linspace(x0, x2, 100)
    y_para = np.polyval(coeffs, x_para)
    
    x_composite = np.concatenate([x_composite, x_para])
    y_composite = np.concatenate([y_composite, y_para])

# Urutkan titik-titik parabola
sorted_idx = np.argsort(x_composite)
x_composite = x_composite[sorted_idx]
y_composite = y_composite[sorted_idx]

# Plot grafik
plt.figure(figsize=(10, 6))
plt.plot(x_smooth, y_smooth, 'b-', label='Fungsi Asli: $f(x) = 2x$')
plt.plot(x_composite, y_composite, 'r--', label='Pendekatan Simpson 1/3')
plt.fill_between(x_composite, y_composite, 0, color='gray', alpha=0.3, label='Daerah Integral')
plt.scatter(x_simp, y_simp, color='red', zorder=5, label='Titik Sampel')
plt.title(f'Integral dengan Metode Simpson 1/3 \nLebar Langkah (h) = {h} \nHasil numberik: {integral:.2f}, nilai sejati: {exact:.2f} \nGalat: {galat:.2f}%')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

# Metode Simpson 3/8
Seperti halnya pada kaidah Simpson 1/3, hampiran nilai integrasi yang lebih teliti
dapat ditingkatkan terus dengan mengunakan polinom interpolasi berderajat lebih
tinggi pula. Misalkan sekarang fungsi f(x) kita hampiri dengan polinom
interpolasi derajat 3. Luas daerah yang dihitung sebagai hampiran nilai integrasi
adalah daerah di bawah kurva polinom derajat 3 tersebut.
Disimpulkan bahwa: <br>
$\displaystyle{\int_{{a}}^{{b}}} f{{\left({x}\right)}}{\left.{d}{x}\right.}={h}{\left[ f{{\left({x}_{{0}}\right)}}+{3} f{{\left({x}_{{1}}\right)}}+{3} f{{\left({x}_{{2}}\right)}}+{2} f{{\left({x}_{{3}}\right)}}+{3} f{{\left({x}_{{4}}\right)}}+{3} f{{\left({x}_{{5}}\right)}}+{2} f{{\left({x}_{{6}}\right)}}+\ldots+ f{{\left({x}_{{n}}\right)}}\right]}$<br>
Dan dapat disederhanakan menjadi:<br>
$\displaystyle{\int_{{a}}^{{b}}} f{{\left({x}\right)}}{\left.{d}{x}\right.}={h}\cdot{\left[ f{{\left({x}_{{0}}\right)}}+{3}\cdot{\sum_{{{i}={i}_{\text{mod}}{3}\ne{0}}}^{{{n}-{1}}}} f{{\left({x}_{{i}}\right)}}+{2}\cdot{\sum_{{{i}={1}_{\text{mod}}{3}={0}}}^{{{n}-{2}}}} f{{\left({x}_{{i}}\right)}}+ f{{\left({x}_{{n}}\right)}}\right]}$

```python
import numpy as np
import matplotlib.pyplot as plt

def Simpson_38(f, a, b, h):
    x = [a]
    i = 0
    
    while x[-1] < b:
        a += h
        if a > b:
            break
        x.append(a)
        
    x = np.array(x)
    y = f(x)
    n = len(x)
    
    hasil = y[0] + y[-1]
    for i in range(1, n-1):
        if i % 3 != 0:
            hasil += 3 * y[i]
        else:
            hasil += 2 * y[i]
    hasil = (3*h / 8) * hasil
    
    return x, y, hasil

# Definisikan fungsi dan parameter
f = lambda x: np.sin(2*x)
a = 0
b = 6
h = 1

x_simp38, y_simp38, integral = Simpson_38(f, a, b, h)
galat = abs((exact - integral)/exact)*100


# Generate kurva halus untuk fungsi asli
x_smooth = np.linspace(a, b, 1000)
y_smooth = f(x_smooth)

# Persiapkan titik-titik untuk visualisasi
x_composite = np.array([])
y_composite = np.array([])

# Fit kubik untuk setiap 3 interval
for i in range(0, len(x_simp38)-1, 3):
    x_seg = x_simp38[i:i+4]  # Ambil 4 titik untuk 3 interval
    y_seg = y_simp38[i:i+4]
    
    # Fit kubik
    coeffs = np.polyfit(x_seg, y_seg, 3)
    
    # Generate titik kubik
    x_para = np.linspace(x_seg[0], x_seg[-1], 100)
    y_para = np.polyval(coeffs, x_para)
    
    x_composite = np.concatenate([x_composite, x_para])
    y_composite = np.concatenate([y_composite, y_para])

# Urutkan titik-titik
sorted_idx = np.argsort(x_composite)
x_composite = x_composite[sorted_idx]
y_composite = y_composite[sorted_idx]

# Plot grafik
plt.figure(figsize=(10, 6))
plt.plot(x_smooth, y_smooth, 'b-', label=f'Fungsi Asli: $f(x) = 2x$')
plt.plot(x_composite, y_composite, 'r--', label='Pendekatan Simpson 3/8')
plt.fill_between(x_composite, y_composite, 0, color='gray', alpha=0.3, label='Daerah Integral')
plt.scatter(x_simp38, y_simp38, color='red', zorder=5, label='Titik Sampel')
plt.title(f'Integral dengan Metode Simpson 3/8 \nLebar Langkah (h) = {h} \nHasil numberik: {integral:.2f}, nilai sejati: {exact:.2f} \nGalat: {galat:.2f}%')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
